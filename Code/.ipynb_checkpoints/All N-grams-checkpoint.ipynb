```python
tqdm.pandas()
def tagger(text):
    doc = nlp (str(text))
    text = ' '.join([chunk.text for chunk in doc.noun_chunks])
    return text

def ner_tagger(text):
    doc = nlp (str(text))
    text = ' '.join([ent.label_ for ent in doc.ents])
    return text
```

```python

allgrams['test'] = allgrams['allgrams'].progress_apply(tagger)
allgrams['test2'] = allgrams['allgrams'].progress_apply(ner_tagger)
```
