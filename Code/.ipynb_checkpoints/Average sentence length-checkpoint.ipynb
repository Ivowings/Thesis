```python
import pandas as pd
import spacy 
from sklearn.utils import shuffle
nlp = spacy.load("en_core_web_lg")
from tqdm import tqdm
#tqdm.pandas() is used for a progress bar on the pandas apply function
tqdm.pandas()
```

```python
df = pd.read_csv("/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/sentences/sentences_emscad.csv")
```

```python
df['length'] = df['sentence'].progress_apply(nlp)
df['length'] = df['length'].progress_apply(len)

average = sum(df['length'])/df.shape[0]

#the average sentence length is 15.17 tokens
print('average sentence length is ',average, ' tokens.')
```
