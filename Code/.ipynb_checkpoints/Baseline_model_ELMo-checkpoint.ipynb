```python
import pandas as pd
import spacy 
nlp = spacy.load("en_core_web_lg")

from tqdm import tqdm
tqdm.pandas()

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import SGDClassifier


from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score

from sklearn.model_selection import learning_curve

```

```python
from simple_elmo import ElmoModel
ELMo = ElmoModel()
ELMo.load("/Users/ivowings/Sync/Thesis/Code/ELMo model wikipedia 2019.zip")
```

```python
path_context = "/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/n-grams context/Annotated/sentence_trigrams_lemmatized_context.csv"
path_trigrams = "/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/n-grams/Annotated/emscad_trigrams_lemmatized_taxonomy.csv"

#Would it be better to keep the context seperated? No context can be combined
df_context = pd.read_csv(path_context,sep=';')
df_context = df_context.dropna()
df_context['label'] = df_context['label'].astype(int)
df_context['allgrams'] = df_context['left_context'] + ' ' + df_context['recovered_gram'] + ' ' + df_context['right_context']
df_context = df_context[['allgrams', 'label']]

df = pd.read_csv(path_trigrams, sep=';')
df = df.dropna()
df['label'] = df['label'].astype(int)

df = pd.concat([df,df_context]).reset_index(drop=True)
```

```python
allgrams = df['allgrams'].tolist()
x = pd.DataFrame(ELMo.get_elmo_vector_average(allgrams,layers='top'))
y = df['label']

seed = 456
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2,random_state=seed)
```

```python
LR = LogisticRegression(solver='newton-cg', random_state=seed)
LR.fit(x_train, y_train)

print('Precision Logistic Regression ',precision_score(y_test, LR.predict(x_test), average='macro'))
print("Recall Logistic Regression ",recall_score(y_test, LR.predict(x_test), average='macro'))
print("F1 score Logistic Regression ",f1_score(y_test, LR.predict(x_test), average='macro'))
```

```python
GBC = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
     max_depth=1, random_state=seed).fit(x_train, y_train) 

print('Precision Gradient Boosting ',precision_score(y_test, GBC.predict(x_test), average='macro'))
print("Recall Gradient Boosting ",recall_score(y_test, GBC.predict(x_test), average='macro'))
print("F1 score Gradient Boosting ",f1_score(y_test, GBC.predict(x_test), average='macro'))
```

```python
SGD = SGDClassifier(loss="hinge", penalty="l2", max_iter=5,random_state=seed)
SGD.fit(x_train, y_train)

print('Precision Stochastic Gradient Descent ',precision_score(y_test, SGD.predict(x_test), average='macro'))
print("Recall Stochastic Gradient Descent ",recall_score(y_test, SGD.predict(x_test), average='macro'))
print("F1 score Stochastic Gradient Descent ",f1_score(y_test, SGD.predict(x_test), average='macro'))
```

```python
RF = RandomForestClassifier(random_state=seed)
RF.fit(x_train, y_train)
RF.score(x_test, y_test)

print('Precision Random Forest ',precision_score(y_test, RF.predict(x_test), average='macro'))
print("Recall Random Forest ",recall_score(y_test, RF.predict(x_test), average='macro'))
print("F1 score Random Forest ",f1_score(y_test, RF.predict(x_test), average='macro'))
```

```python
def training_graph(model,x,y,x_axis_size, name):
    
    precision_scores=[]
    recall_scores=[]
    f1_scores=[]
    training_size=[]

    for training_amount in range(x_axis_size,(len(x)+x_axis_size),x_axis_size):

        X = x[0:training_amount]
        Y = y[0:training_amount]

        x_train,x_test,y_train,y_test=train_test_split(X, Y, test_size=0.2,random_state=seed)

        model.fit(x_train, y_train)

        precision_scores.append(precision_score(y_test, model.predict(x_test), average='macro'))
        recall_scores.append(recall_score(y_test, model.predict(x_test), average='macro'))
        f1_scores.append(f1_score(y_test, model.predict(x_test), average='macro'))
        training_size.append(training_amount)
        
    
    plt.plot(training_size,precision_scores)
    plt.plot(training_size,recall_scores)
    plt.plot(training_size,f1_scores)
    
    plt.ylabel('Precision')
    plt.xlabel('Training size')
    plt.title('Training graph {}'.format(name))
    
    plt.legend(['Precision', 'Recall', 'F1'])
    plt.show()
```

```python
training_graph(LR,x,y,50, 'LR')
training_graph(GBC,x,y,50, 'GBC')
training_graph(SGD,x,y,50, 'SGD')
training_graph(RF,x,y,50, 'RF')
```
