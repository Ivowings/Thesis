```python
import pandas as pd
```

```python
df_hardskills = pd.read_csv("/Users/ivowings/Sync/Thesis/Datasources/Skills/emsi_skills_ner.csv",sep=';')
df_hardskills = df_hardskills[df_hardskills.type == 'Hard Skill']
df_hardskills = df_hardskills[['name', 'ner']]
df_hardskills

```

```python
df_hardskills['name'] = df_hardskills['name'].str.replace(r'[^A-Za-z0-9 ]+', ' ')
df_hardskills['name'] = df_hardskills['name'].str.replace(r'  ', ' ')
df_hardskills['name'] = df_hardskills['name'].str.strip()
df_hardskills['name'] = df_hardskills['name'].str.lower()
df_hardskills
```

```python
df_hardskills.to_csv('/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/Skills/emsi_skills.csv',index=False)
```
