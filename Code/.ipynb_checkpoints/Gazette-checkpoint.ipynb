```python
grams = grams['concatenated'].str.split('|',expand=True)
grams[1] = grams[1].str.strip()

grams
```

```python
df_hardskills = pd.read_csv("/Users/ivowings/Sync/Thesis/Datasources/Preprocessed/Skills/emsi_skills.csv")
df_hardskills
```

```python
#Initializing the keyword processor
keyword_processor = KeywordProcessor(case_sensitive=False)

#Adding all the skills to the processor
for skill in tqdm(df_hardskills['name'].str.strip()):
    keyword_processor.add_keyword(skill)

def searcher(row):
    #check if the words are in the row and return a True or False instead of the actual word
    boolean = bool(keyword_processor.extract_keywords(row))
    return boolean

def index_match_start(row):
    try:
        match = keyword_processor.extract_keywords(row,span_info=True)
        index = [index[1] for index in match][0]
        return index
    except:
        None

def index_match_end(row):
    try:
        match = keyword_processor.extract_keywords(row,span_info=True)
        index = [index[2] for index in match][0]
        return index
    except:
        None

grams['match'] = grams[1].progress_apply(searcher)
grams['index_start'] = grams[1].progress_apply(index_match_start)
grams['index_end'] = grams[1].progress_apply(index_match_end)
grams['letters'] = grams[1].progress_apply(len)
```
