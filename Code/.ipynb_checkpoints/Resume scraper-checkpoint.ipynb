```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
from bs4 import BeautifulSoup
import csv
import time
import pandas as pd
```

```python
pagecounter = 202 #ready to be run again
results = []
while len(results) <= 100000: 
    
    session = requests.Session()
    retry = Retry(total=10000, connect=10000, backoff_factor=1)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    
    
    url = "https://www.postjobfree.com/resumes?q=&l=United+States&radius=25&r=100&p=" + str(pagecounter)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title_tags = soup.find_all('h3', attrs={'class': 'itemTitle'})
    links = ["https://www.postjobfree.com" + title_tag.a['href'] for title_tag in title_tags]
    for link in links:
        res = requests.get(link)
        print(res.status_code, link)
        content = BeautifulSoup(res.content, 'html.parser')
        try:
            results.append({
                'job_title': content.find('div', attrs={'class': 'leftColumn'}).find('h1').get_text(),
                'posted_date': content.find('span', attrs={'class': 'colorDate'}).get_text(),
                'resume': content.find('div', attrs={'class': 'normalText'}).get_text()[:-23]
            })
        except:
            continue
        #time.sleep(3)
    print('number of resumes', len(results))
    pagecounter = pagecounter + 1

```

```python
df = pd.DataFrame(columns=['job_title','posted_date','resume'])
```

```python
for row in results:
    df = df.append(row,ignore_index=True)
```

```python
df.to_csv('/Users/ivowings/Thesis/Code/postjobfree_resumes_US_6.csv',sep=';',header=True,index=True)
```
