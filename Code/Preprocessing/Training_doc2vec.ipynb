```python
import pandas as pd
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from tqdm import tqdm
from bs4 import BeautifulSoup
from nltk import tokenize
from nltk import word_tokenize
from gensim.models import Phrases
```

```python
emscad = pd.read_csv("/Users/ivowings/Sync/Thesis/Datasources/Job vacancies/emscad_v1 2.csv")

#Removing the html tags in the columns.
rows = []
for t in tqdm(emscad['description']):
    soup = BeautifulSoup(t,"lxml")
    rows.append(soup.get_text())
emscad['description'] = rows

rows = []
#Some columns contain nan therefore appending no text
for t in tqdm(emscad['requirements']):
    if str(t) == 'nan':
        rows.append('')
    else:
        soup = BeautifulSoup(t,"lxml")
        rows.append(soup.get_text())
emscad['requirements'] = rows

emscad['job_description'] = emscad['description'] + ' ' + emscad['requirements']
df = emscad[['job_description']]
```

```python
reed = pd.read_csv('/Users/ivowings/Sync/Thesis/Datasources/Job vacancies/reed_uk.csv')
reed['job_requirements'] = reed['job_requirements'].fillna('')
reed['job_description'] = reed['job_description'] + ' ' + reed['job_requirements']
df2 = reed[['job_description']]
df = pd.concat([df,df2])
```

```python
tqdm.pandas()
sentences = []

#Breaking the job postings into sentences

for vacancy in tqdm(df['job_description']):
    sentences_storage = tokenize.sent_tokenize(vacancy)
    for sentence in sentences_storage:
        if sentence != '':
            sentences.append(sentence)
    
    
```

```python
from nltk.tokenize import RegexpTokenizer

#Removing the punctiation from the sentences
tokenizer = RegexpTokenizer(r'\w+')
sentences_cleaned = []
for sentence in tqdm(sentences):
    tokens = tokenizer.tokenize(sentence)
    sentences_cleaned.append(tokens)
```

```python
#Lowercasing all the text
sentences_cleaned_final = []
for sentence in tqdm(sentences_cleaned):
    sublist = []
    for token in sentence:
        sublist.append(token.lower())
    sentences_cleaned_final.append(sublist)
        #sentences_cleaned_final.append(token.lower())
        
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(sentences_cleaned_final)]
```

```python
%%time
#dm ({1,0}, optional) – Defines the training algorithm. 
#If dm=1, ‘distributed memory’ (PV-DM) is used. Otherwise, distributed bag of words (PV-DBOW) is employed.
model = Doc2Vec(documents, vector_size=100, window=2, min_count=1, workers=4, dm=1)
model.save("/Users/ivowings/Sync/Thesis/Code/Doc2vec models/Doc2vec_s100_pvdm.model")
```
